```python
def auto_quantize(bdata,colorcount = 6):
    """
    return indexed colors, color_idx
    """
    img = bdata.astype(np.float64)
    # reshaping the pixels matrix
    pixel = np.reshape(img,(img.shape[0]*img.shape[1],3))

    # performing the clustering
    centroids,_ = kmeans(pixel,colorcount) # up six colors will be found
    # quantization
    qnt,_ = vq(pixel,centroids)

    # reshaping the result of the quantization
    centers_idx = np.reshape(qnt,(img.shape[0],img.shape[1]))
    return centers_idx,centroids
```

```python
def coltoBspace(col,colType,ini):

    for rowidx,nodeCol in enumerate(col):
        
        if colType == 0:
            yield ini + [rowidx,rowidx,nodeCol]
        else:
            yield ini + [rowidx,rowidx+.5,nodeCol]
            
            
def nodesOf(bdata):
    nodes = []

    for colidx in range(bdata.shape[1]):
        
        colType = colidx%2
        if colType == 0:
            ini = [colType,colidx,colidx]
            col = bdata[:,colidx]

            nodes.extend(list(coltoBspace(col,colType,ini)))
        else:
            ini = [colType,colidx,colidx]
            col =bdata[:-1,colidx]
            nodes.extend(list(coltoBspace(col,colType,ini)))
    "colidx","x","rowidx","y","cidx"
    return pd.DataFrame(nodes,columns=["coltype","colidx","x","rowidx","y","cidx"])
```

```python
def getAllInputs(aucc,colorCount,targetdist,colidx):
    """
    color constraints
    colorCount
    targetminimalDistibution
    
    all in U color space
    
    """
    # for colidx = 0 2 4 
    colorOps = list(range(colorCount))

    # build many inputs 
    inputs = []
    for oneNodeColor in aucc:
        possibleIn = [(lc,rc) for lc,rc in itertools.product(colorOps,colorOps) if (lc == oneNodeColor or rc == oneNodeColor)]
        inputs.append(possibleIn)
    inputs = (functools.reduce(operator.concat, _) for _ in itertools.product(*inputs))

    inputs= list(inputs)
    if colidx %2 == 1:
        wops = [(_,) for _ in colorOps]
        inputs = [functools.reduce(operator.concat, _) for _ in itertools.product(wops, inputs,wops)]


    # filter possible input based on distibution 
    filtered=[]
    for inp in inputs:
        idist = Counter(inp)
        okdist =True
        for k,v in targetdist.items():
            if idist[k]<v:
                okdist = False
                break
        if okdist:
            filtered.append(inp)
    return filtered
```

```python

def nodeOptions(lc,rc,cc):
    """
    left and right color 
    color constraint
    """
    if lc == rc :
        return [('RR' , (lc,rc))]
    else:
        if lc == cc:
            return [('RR' , (rc,lc)),
            ('RL',  (lc,rc))]
        elif rc == cc:
            return [('LL' , (rc,lc)),
            ('LR',  (lc,rc))]
        
def possibleNoding(colidx,oneInput,aucc):
    """
    # calculate every noding option (colidx,oneInput,aucc)



    colidx,
    one possible input
    colorConstraints
    """
    
    if colidx%2 == 1:
        coloptions = []
        for i in range(len(aucc)):
            cc = aucc[i]

            lc,rc = oneInput[i*2+1:i*2+3]
            coloptions.append(nodeOptions(lc,rc,cc))
        possibleNoding = list(itertools.product(*coloptions))
        
        aa = []
        for onestage in possibleNoding:
            noding = tuple([nodeType for nodeType,(lc,rc) in onestage])
            outputw = functools.reduce(operator.concat, [(lc,rc) for nodeType,(lc,rc) in onestage])
            outputw = (oneInput[0],)+outputw+( oneInput[-1],)
            aa.append( (colidx,oneInput,noding,outputw))

        possibleNoding = aa

        return possibleNoding
    
    
        
    else:
        coloptions = []
        for i in range(len(aucc)):
            cc = aucc[i]

            lc,rc = oneInput[i*2:i*2+2]
            coloptions.append(nodeOptions(lc,rc,cc))

        possibleNoding = list(itertools.product(*coloptions))


        aa = []
        for onestage in possibleNoding:
            noding = tuple([nodeType for nodeType,(lc,rc) in onestage])
            outputw = functools.reduce(operator.concat, [(lc,rc) for nodeType,(lc,rc) in onestage])
            aa.append( (colidx,oneInput,noding,outputw))

        possibleNoding = aa

        return possibleNoding
```

```python

def findNoding(colidx,colorConstraints,uminDists,colorCount):

    # get color constraint as tuple
    aucc =colorConstraints["ucc"][colidx]


    # get target minimal distribution
    targetdist  = uminDists[colidx]




    # build all inputs that match color constraints & targetDist
    inputs = getAllInputs(aucc,colorCount,targetdist,colidx)
    

    # find all possible noding with those inputs
    res = []
    for oneInput in inputs:
        res+=possibleNoding(colidx,oneInput,aucc)


    cols = ["colidx","oneInput","noding","outputw"]

    dfres = pd.DataFrame(res,columns=cols)
    return dfres
```
